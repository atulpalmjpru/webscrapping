#Importing and loading the urls

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

#Extraction  of the Desired Data of particular company extracted in below method

```python

def Information(symbol):
  url="https://finance.yahoo.com/quote/"+symbol+"?p="+symbol
  res=req.get(url)
  ls_name=[]
  ls_value=[]
  if res.status_code==200:
    try:
      t=res.text
      soup=bs(t,features="html.parser")
      name=soup.find_all("span",class_='label svelte-tx3nkj')
      value=soup.find_all("span",class_='value svelte-tx3nkj')
      for i in range(len(name)):
        ls_name.append(name[i].text)
        ls_value.append(value[i].text)
    except Exception as e:
      pass
  return ls_name,ls_value
```

#Extraction of the name of top 50 companies from wikipedia

```python
def Company_name():
  wikiUrl="https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
  response=req.get(wikiUrl)
  st_code=response.status_code
  symbols=[]
  if st_code==200:
    page=response.text
    #print(page)
    beautify_page=bs(page,features="html.parser")

    trs1=beautify_page.find_all("tbody")[0]
    #print(len(trs1))
    #print(trs1.contents[4].contents[1].text)
    for i in range(len(trs1)):
      if i<2:
        continue
      if i%2==0:
        symbols.append(trs1.contents[i].contents[1].text.strip("\n"))
    return symbols
symbols=Company_name()
data={"symbol":[],
      "metric":[],
      "value":[]}
print(symbols)



```

```python
for symbol in symbols:
  name,value=Information(symbol)
  for i in range(len(name)):
    if symbol not in data:
      data["symbol"].append(symbol)
      data["metric"].append(name[i])
      data["value"].append(value[i])
print(data)

data1=pd.DataFrame(data)
data1
```

#Importing to CSV file-- Output as csv file

```python
df=pd.DataFrame(data)
df.to_csv("financial_Data_project.csv")
```
